```python
datos["Hombres"]=datos["Hombres"].fillna(0)
datos["Mujeres"]=datos["Mujeres"].fillna(0)
datos
```

```python
def filtro_res(Categoria):
    return "Respiratoria" in Categoria 

datosRES = datos[datos['Categoria'].apply(filtro_res)]

infoRES = datosRES[['Enfermedad',"Hombres", "Mujeres"]]
infoRES
```

```python
datosRES.info()
```

```python
res_h = datosRES['Hombres']
prom_h = res_h.mean()
prom_h
```

```python
res_m = datosRES['Mujeres']
prom_m = res_m.mean()
prom_m
```

```python
compar_data = pd.DataFrame({'Sexo':['Hombres','Mujeres'], 'Promedio':[prom_h,prom_m]})
compar_data
```

```python
plt.pie(compar_data["Promedio"],labels=compar_data['Sexo'], autopct='%1.1f%%')
plt.title('Afectación de enfermedades respiratorias')
plt.show
```
